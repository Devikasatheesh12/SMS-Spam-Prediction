```python
import pandas as pd
df1=pd.read_csv("/content/drive/MyDrive/cnn/dataset/Copy of SMS_train.csv",encoding='ISO-8859-1')
df1 #utf-8 doest no supprt here.so use other encoding scheme
```

```python
df2=pd.read_csv("/content/drive/MyDrive/cnn/dataset/Copy of SMS_test.csv",encoding='ISO-8859-1')
df2
```

```python
df=pd.concat([df1,df2],ignore_index=True)
df
```

```python
df.drop(['S. No.'],axis=1,inplace=True)
```

```python
df
```

```python
df['Label']=df['Label'].map({'Non-Spam':0,'Spam':1})
df
```

```python
msg=df.Message_body
msg
```

```python
msg=msg.str.replace("[^a-zA-Z0-9 ]","") #removing special character,replacing it with a space
msg
```

```python
import nltk
from nltk.stem import SnowballStemmer
nltk.download('punkt')
stemmer=SnowballStemmer('english')
from nltk.tokenize import word_tokenize
```

```python
msg=msg.apply(lambda row:[stemmer.stem(token.lower()) for token in word_tokenize(row)]).apply(lambda token:" ".join(token))
msg
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
sw=stopwords.words('english')
sw
```

```python
msg=msg.apply(lambda row:[token for token in word_tokenize(row) if token not in sw]).apply(lambda token:" ".join(token))
msg
```

```python
msg=msg.apply(lambda row:[token for token in word_tokenize(row) if len(token)>2]).apply(lambda token:" ".join(token))
msg
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
vect=vectorizer.fit_transform(msg)
print(vect)
```

```python
vect.shape
```

```python
y=df.Label
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(vect,y,test_size=0.3,random_state=1)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
nb.fit(X_train,y_train)
```

```python
y_pred=nb.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(X_train,y_train)
```

```python
y_pred1=sv.predict(X_test)
```

```python
print(classification_report(y_test,y_pred1))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
y_pred2=rf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred2))
```

```python
from sklearn.ensemble import AdaBoostClassifier
ad=AdaBoostClassifier()
ad.fit(X_train,y_train)
```

```python
y_pred3=ad.predict(X_test)
```

```python
print(classification_report(y_test,y_pred3))
```
